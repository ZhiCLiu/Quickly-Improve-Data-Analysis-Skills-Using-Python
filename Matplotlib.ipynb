## Subplot

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1)

for ax in axes:
    ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1)

axes[0].plot(x,y)
fig.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)
fig.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)
fig.subplots_adjust()#fail!
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_title('first')
axes[0][1].plot(x,y)
axes[0][1].set_ylabel('2Y')
axes[1][0].plot(x,y)
axes[1][0].set_xlabel('3X')
axes[1][1].plot(a,b)
fig.suptitle('Figure Level',fontsize = 16)
fig.subplots_adjust(wspace = 1,hspace =0.5)
```

## Styling

```python
x = np.linspace(0,10,11)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label = 'X vs X')
ax.plot(x,x**2,label = 'X vs X^2')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label = 'X vs X')
ax.plot(x,x**2,label = 'X vs X^2')
ax.legend(loc = 'upper left')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label = 'X vs X')
ax.plot(x,x**2,label = 'X vs X^2')
ax.legend(loc = (1,0.5))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color = 'lightpink',label = 'x v x',lw=5,ls='-.') #color="#8B008B"# RGB hex code
ax.plot(x,x+1,color = 'lightblue',label = 'x vs x+1',lw=2,ls='--') 
ax.plot(x,x+2,color = 'lightgrey',label = 'x vs x+1',lw=2,ls=':') 
ax.legend()
```
